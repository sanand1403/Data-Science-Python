```python
# Home Loan EMI Calculator

def calculate_emi(principal, rate, years):
    # Monthly interest rate
    monthly_rate = rate / (12 * 100)
    # Total number of months
    months = years * 12
    
    # EMI formula
    emi = (principal * monthly_rate * (1 + monthly_rate) ** months) / ((1 + monthly_rate) ** months - 1)
    
    # Total payment & interest
    total_payment = emi * months
    total_interest = total_payment - principal
    
    return round(emi), round(total_interest), round(total_payment)


# ---- Input Section ----
print("-- Input Section --")
principal = float(input("Loan Amount (₹): "))
years = int(input("Tenure (Years): "))
rate = float(input("Interest Rate (% P.A.): "))

# ---- Calculation ----
emi, interest, total = calculate_emi(principal, rate, years)

# ---- Output Section ----
print("\n-- Output Section --")
print(f"Monthly Home Loan EMI : ₹{emi:,}")
print("View Details")
print(f"Principal Amount      : ₹{principal:,.0f}")
print(f"Interest Amount       : ₹{interest:,}")
print(f"Total Amount Payable  : ₹{total:,}")
```
