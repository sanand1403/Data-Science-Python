```python
# import lib
import pandas as pd
import matplotlib.pyplot as plt
import yfinance as yf
```

```python
# download stock data 
data = yf.download('GOOG', start='2020-01-01', end='2023-12-31')
data.head() # print top 5 rows
```

```python
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['Close'], label='GOOG Close Price')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.title('Google Inc. Stock Price (2020-2023)')
plt.legend()
plt.grid()
plt.show()
```

```python
# moving average (Trend Detection)

data["MA50"] = data["Close"].rolling(window=50).mean()
plt.figure(figsize=(12, 6))
plt.plot(data.index, data['MA50'], label='50-Day MA', color='orange')
plt.plot(data.index, data['Close'], label='GOOG Close Price', color='blue')
plt.xlabel('Date')
plt.ylabel('Price (USD)')
plt.title('Google Inc. Stock Price with 50-Day Moving Average')
plt.legend()
plt.grid()
plt.show()
```
