```python
# Program to find second largest number

# Example 1
numbers = [10,20,34,55,28,12,50,79,94,88]
unique_numbers = list(set(numbers))
second_largest = sorted(unique_numbers)[-2]
print("Second largest number is:", second_largest)
```
