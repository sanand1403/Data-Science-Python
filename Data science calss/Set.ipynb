```python
# Sets
# Set Union (Adds elements from both sets)
A = {1,2,3}
B = {2,3,4,5}
result = A|B
result1 = B.union(A)
print(result)
print(result1)
```

```python
# Set Intersection (Shows common elements from both sets)
A = {1,2,3}
B = {2,3,4,5}
result = A & B
print(result)
result1 = B.intersection(A)
print(result1)
```

```python
# Set Difference (Shows uncommon elements in the first set)
A = {1,2,3}
B = {2,3,4,5}
result = A-B
print(result)
result1 = B-A
print(result1)
result2 = A.difference(B)
print(result2)
result3 = B.difference(A)
print(result3)
```

```python
# Set Symmetric Difference (Shows Uncommon elements in both sets)
A = {1,2,3}
B = {2,3,4,5}
result = A ^ B
print(result)
result1 = A.symmetric_difference(B)
print(result1)
```
