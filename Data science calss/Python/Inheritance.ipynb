```python
# Step 1 define the parent class
class Animal:
    def __init__(self,name,species):
        self.name = name
        self.species = species
    def make_sound(self):
        print('Some generic animal sound')       
```

```python
# step 2 Define the child class
class Dog(Animal): # Dog inherits from Animal
    def __init__(self, name, breed):
        super().__init__(name, species="Dog")
        self.breed = breed

    def make_sound(self):
        print(f'{self.name} is barks : woof !...')
    def fetch(self):
        print(f'{self.name} is fetching the ball')
```

```python
# Animal object
generic_animal = Animal("leo","Lion")
generic_animal.make_sound()

# dog object
my_dog = Dog("Buddy", "Golden Retriver")
my_dog.make_sound()

my_dog.fetch()

print(f'name: {my_dog.name} species : {my_dog.species} breed : {my_dog.breed}')
```

```python
class Cat(Animal):
    def __init__(self, name, color):
        super().__init__(name, species="Cat")
        self.Color = color

    def make_sound(self):
        print(f'{self.name} is meows : Meow !....')
    def Scratch(self):
        print(f'{self.name} is Scratching Furniture')
```

```python
my_cat = Cat("Whiskers" , "Gray")
my_cat.make_sound()
my_cat.Scratch()
print(f'Name : {my_cat.name} species : {my_cat.species} color : {my_cat.Color}')
```

```python
# 2nd Example Vehicles
# Step 1: Define the parent class
class Vehicle:
    def __init__(self, name, vehicle_type):
        self.name = name
        self.vehicle_type = vehicle_type
    
    def move(self):
        print("The vehicle is moving...")


# Step 2: Define the child class Car
class Car(Vehicle):  # Car inherits from Vehicle
    def __init__(self, name, brand):
        super().__init__(name, vehicle_type="Car")
        self.brand = brand

    def move(self):
        print(f"{self.name} (Car) is driving on the road 🚗")
    
    def honk(self):
        print(f"{self.name} is honking: Beep! Beep!")


# Step 3: Define another child class Bike
class Bike(Vehicle):
    def __init__(self, name, bike_type):
        super().__init__(name, vehicle_type="Bike")
        self.bike_type = bike_type

    def move(self):
        print(f"{self.name} (Bike) is zooming on the street 🏍️")
    
    def do_wheelie(self):
        print(f"{self.name} is doing a wheelie!")


# Step 4: Define another child class Truck
class Truck(Vehicle):
    def __init__(self, name, capacity):
        super().__init__(name, vehicle_type="Truck")
        self.capacity = capacity

    def move(self):
        print(f"{self.name} (Truck) is hauling heavy loads 🚚")
    
    def load_cargo(self):
        print(f"{self.name} is loading {self.capacity} tons of cargo.")


# --- Testing the code ---

# Generic vehicle
generic_vehicle = Vehicle("Transporter", "Unknown")
generic_vehicle.move()

# Car object
my_car = Car("Speedster", "Tesla")
my_car.move()
my_car.honk()
print(f"Name: {my_car.name}, Type: {my_car.vehicle_type}, Brand: {my_car.brand}")

# Bike object
my_bike = Bike("Thunderbolt", "Sports Bike")
my_bike.move()
my_bike.do_wheelie()
print(f"Name: {my_bike.name}, Type: {my_bike.vehicle_type}, Bike Type: {my_bike.bike_type}")

# Truck object
my_truck = Truck("BigHauler", 15)
my_truck.move()
my_truck.load_cargo()
print(f"Name: {my_truck.name}, Type: {my_truck.vehicle_type}, Capacity: {my_truck.capacity} tons")
```
