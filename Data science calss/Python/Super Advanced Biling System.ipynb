```python
# Initialize lists to store products, prices, and quantities
products = []
prices = []
quantities = []

# User input section
count = 1
while True:
    product = input(f"Enter Product {count} -\n")
    products.append(product)
    
    price = int(input(f"Enter {product} Price :\n").replace("/-", ""))
    prices.append(price)
    
    quantity = int(input(f"Enter {product} QT :\n"))
    quantities.append(quantity)
    
    print("-----------------------------")
    more = input("Do you Want to add more [Y/N]\n").upper()
    if more != 'Y':
        break
    count += 1

# Calculate total for each product
total_per_product = [price * qty for price, qty in zip(prices, quantities)]
total = sum(total_per_product)

# GST
add_gst = input("\nDo You Want to add GST [Y/N]\n").upper()
gst_percent = 0
if add_gst == 'Y':
    gst_percent = float(input("Enter GST % = "))

# Offer
print("\nSelect Offer :")
print("1. FLAT 2% OFF")
print("2. FLAT 4% OFF")
print("3. FLAT 8% OFF")
offer_choice = input()
offer_dict = {'1': 2, '2': 4, '3': 8}
offer_percent = offer_dict.get(offer_choice, 0)

# Print bill
print("\n-------Output-------\n")
for i, amount in enumerate(total_per_product, start=1):
    print(f"{i}. {products[i-1]} = {amount}/-")

print("-------------------")
print(f"Total = {total}")

gst_amount = 0
if gst_percent > 0:
    gst_amount = total * gst_percent / 100
    print(f"GST = {gst_percent}%")

offer_amount = 0
if offer_percent > 0:
    print(f"offer = {offer_percent}%")
    offer_amount = (total + gst_amount) * offer_percent / 100

final_total = total + gst_amount - offer_amount
print("-------------------")
print(f"Final Total = {final_total}/-")
```
