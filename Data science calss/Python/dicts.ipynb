```python
# name = {key/;value,key:value}

data = {"name": "Joy","age":23,"height": 4.5}

print(data)
```

```python
# access single value
print(data['name'])
```

```python
# update data
data['age'] = 54
print(data)
```

```python
# delete
data.pop("age")
print(data)
```

```python
# add new item into dict...
data['city']="mumbai"
print(data)
```

```python
# keys(): Returns a list of all keys in the dictionary.
print(data.keys())
# values(): Returns a list of all values in the dictionary.
print(data.values())
# items(): Returns a list of key-values pairs as tuples.
print(data.items())
# get(key, default):Returns the value for a key if it exists,else returns a default value.
print(data.get("name"))
# pop(key,default): Removes and returns the value associated with the key.
```

```python
# print using loop

for i in data:
    print(i,'=',data[i])
```
