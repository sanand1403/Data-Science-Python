```python
# Product Inventory Program

def main():
    inventory = {}

    while True:
        #print("\n--- Input Section ---")
        sku = input("Enter SKU ( Unique Number ) : ").strip()
        product_name = input("Enter Product Name : ").strip()
        quantity = int(input("Enter Product QT : "))
        price = int(input("Enter Single Product Price (without /-) : "))

        # Store product in dictionary with SKU as key
        inventory[sku] = {
            "name": product_name,
            "quantity": quantity,
            "price": price
        }

        choice = input("\nDo you Want Add more [Y/N] : ").strip().upper()
        if choice == "N":
            break

    # Output Section
    print("\n--- Output Section ---")
    print("------ Search Item ------")
    search_sku = input("Enter Product Item From SKU No. : ").strip()

    if search_sku in inventory:
        product = inventory[search_sku]
        total_cost = product["quantity"] * product["price"]

        print(f"Total Cost : {total_cost}/-")
        print(f"Product Name : {product['name']}")
        print(f"QT : {product['quantity']}")
        print(f"Single Price : {product['price']}/-")
    else:
        print("❌ Product not found in inventory.")

if __name__ == "__main__":
    main()
```
