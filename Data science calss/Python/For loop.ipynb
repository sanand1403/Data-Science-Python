```python
# how to take user input

start = int(input('Enter start number'))
end = int(input('Enter end number'))

for i in range(start,end+1):
    print (i,end=' ')
```

```python
# odd and even number

a = int(input("Enter number"))
for i in range(a):
    if i % 2 == 0:
       print(i,"is even")
    else:
       print(i,"is odd")
```

```python
# 1 to 10 total

Start = int(input("Enter start number"))
End = int(input("Enter end number"))
result = sum(range(Start,End+1))
print('The sum is',result)
```
