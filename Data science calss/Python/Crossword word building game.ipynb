```python
# Crossword Word Building Game with Scoring

print("Crossword Word Building Game")
print("Question:")
print("What is the national animal of India?")

# Step 1: Take maximum character length from user
max_length = int(input("Enter the maximum character length of the answer: "))

# Step 2: Input the answer character by character
answer_chars = []
for i in range(1, max_length + 1):
    char = input(f"Character {i} : ").upper()   # store in uppercase for consistency
    answer_chars.append(char)

# Step 3: Join characters to form the word
answer = "".join(answer_chars)

# Step 4: Display the final answer
print("Answer -", answer.capitalize())

# Step 5: Initialize game
current_word = answer.capitalize()
score = 0

while True:
    print(f"\nCurrent Word: {current_word}")
    chosen_char = input(f"Choose a character from '{current_word}': ").upper()

    # check if chosen char exists in word
    if chosen_char not in current_word.upper():
        print(f"'{chosen_char}' is not in '{current_word}'. Try again.")
        continue

    new_word = input(f"Make a new word using '{chosen_char}': ").capitalize()
    score += 10  # add points for valid word
    print(f"✅ New Word: {new_word} (+10 points)")
    print(f"🏆 Current Score: {score}")

    choice = input("Do you want to continue? (Yes/No): ").strip().lower()
    if choice != "yes":
        print("\n🎮 Game Over!")
        print(f"Your Final Score: {score}")
        break

    current_word = new_word  # update word for next round
```
