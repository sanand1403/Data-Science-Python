```python
# Advanced Billing System
# Initialize lists to store products and their prices
products = []
prices = []

# User input section
count = 1
while True:
    product = input(f"Enter Product {count} -\n")
    products.append(product)
    
    more = input("Do you Want to add more [Y/N]\n")
    if more.upper() != 'Y':
        break
    count += 1

# Enter prices for each product
for product in products:
    price = int(input(f"Enter {product} Price :\n").replace("/-", ""))
    prices.append(price)

# Ask for GST
add_gst = input("\nDo You Want to add GST [Y/N]\n").upper()
gst_percent = 0
if add_gst == 'Y':
    gst_percent = float(input("Enter GST % = "))

# Print bill
print("\n-------Output-------\n")
total = 0
for i, (product, price) in enumerate(zip(products, prices), start=1):
    print(f"{i}. {product} = {price}/-")
    total += price

print("-------------------")
print(f"Total = {total}")

if gst_percent > 0:
    gst_amount = total * gst_percent / 100
    print(f"GST = {gst_percent}%")
    print("-------------------")
    final_total = total + gst_amount
    print(f"Final Total = {int(final_total)}/-")
else:
    print("-------------------")
    print(f"Final Total = {total}/-")

```
