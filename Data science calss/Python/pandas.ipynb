```python
# pd.DataFrame

students ={
    "name":['Anand','Mihir','Rupesh','Srikanth','Arvind'],
    "age":[29,25,24,24,25],
    "city":['mumbai','pune','delhi','chennai','bangalore']
}

df = pd.DataFrame(students)
print(df)

# change index

indexx=[101,102,103,104,105]

df = pd.DataFrame(students,index=indexx)
print(df)

df.to_csv('student.csv',index=False)
```
