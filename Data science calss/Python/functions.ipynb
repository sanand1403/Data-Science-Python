```python
def mrp(price,gst,quantity,discount):
    GSTAmount = price*gst/100
    FinalAmount = GSTAmount+price
    TotalAmount = FinalAmount*quantity
    Finalprice = TotalAmount-TotalAmount*discount/100
    print('gst :',gst,'%')
    print('GSTAmount :',GSTAmount,'₹')
    print('FinalAmount :',FinalAmount,'₹')
    print('TotalAmount :',TotalAmount,'₹')
    print('Finalprice :',Finalprice,'₹')

```

```python
mrp(1000,12,5,10)
```
