```python
# User input List

Food = []
Price = []
size = int(input('Enter List size...'))
for i in range(size):
    i = input('Enter Food Name...')
    Food.append(i)

for i in range(size):
    i = int(input('Enter price...'))
    Price.append(i)

for i in range(size):
    print(Food[i],' = ',Price[i])

print('-----------------------------')
Total_price = sum(Price)
print("Total Price = ", Total_price)
Gst = (Total_price*18/100)
print("Gst = ", Gst)

print('-----------------------------')
Final_amount = Total_price + Gst
print("Final Amount = ", Final_amount)
```
