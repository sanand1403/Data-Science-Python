```python
# Encapsulation
class Student:
    def __init__(self,name,age,marks):   # Constructor
        self.name = name
        self.age = age
        self.marks = marks
    def show_details(self):
        print(f'Name : {self.name}')
        print(f'Age : {self.age}')
        print(f'Marks : {self.marks}')
    def calculate_grade(self):
        if self.marks >100:
            return "Please enter marks only between 0 and 100"
        elif self.marks>=90:
            return "S"    
        elif self.marks >=75:
           return "A"
        elif self.marks >=60:
           return "B"
        elif self.marks >=40:
          return "C"
        else :
          return "Fail"
    def get_grade(self):
       return self.calculate_grade()
```

```python
s=Student("Anand",29,60)
s.show_details()
print(f'Grade : {s.get_grade()}')
```

```python
class Product:
    def __init__(self,name,price,gst,discount):   # Constructor
        self.name = name
        self.price = price
        self.gst = gst
        self.discount = discount
    def show_details(self):
        print(f'Name : {self.name}')
        print(f'Price : {self.price}')
        print(f'GST Rate(%) : {self.gst}')
        print(f'Discount : {self.discount}')
    def calculate_gst(self):
        self.gst=(self.price*self.gst)/100
        print(f'The GST on product price is {self.gst}')
        
    def calculate_discount(self):
        self.discount=(self.price*self.discount)/100
        print(f'The Discount on product price is {self.discount}')
    
    def final_price(self):
       Finale_Price = self.price + (self.gst) - (self.discount)
       print(f'The Final Price icluding GST and Discount is {Finale_Price}')
```

```python
p=Product('Burger',1000,18,10)
p.show_details()
p.calculate_gst()
p.calculate_discount()
p.final_price()
```
