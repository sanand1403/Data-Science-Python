```python
class Bank:
    def check_balance(self,amount):
        self.amount = amount
        print(f'Bank Main Balance = {self.amount}')
    def deposit(self,deposit_amount):
        self.deposit = deposit_amount
        self.amount = self.amount + self.deposit
        print (f'deposit = {self.deposit}')
        print(f'Bank Main Balance = {self.amount}')
    def withdraw(self,withdraw_amount):
        if withdraw_amount>self.amount:
            print("Insufficient Balance")
        else :
            self.withdraw = withdraw_amount
            self.amount=self.amount-self.withdraw
            print(f'withdraw = {self.withdraw}')
            print(f'Bank Balance = {self.amount}')
```

```python
bank = Bank()
bank.check_balance(50000)
bank.withdraw(5000)
bank.deposit(500)
bank.withdraw(20000)
```
