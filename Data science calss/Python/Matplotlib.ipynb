```python
import matplotlib.pyplot as plt

# sample data

x = [2,3,4,5,6] # x axis value
y = [7,8,9,10,11] # y axis value
z = [10,20,30,40,50] # z axis value

plt.plot(y,x, color='red', linestyle='--' , marker='o',markersize=8)
plt.plot(y,z, color='yellow',linestyle='-', marker='o',markersize=8)

# add labels and title
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Simple line plot')
plt.show()
```

```python
cat = ['java','C++','python','go']
value = [20,30,40,10]

plt.bar(cat,value,color='purple')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Bar Graph')
plt.show()
```

```python
# adding grid
plt.plot(x,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Grid Plot')
plt.grid(True)
plt.show()
```

```python
# scatter plot

plt.scatter(x,y)
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Scatter Plot')
plt.show()
```

```python
# Histogram plot

import numpy as np
np.random.seed(43)
data = np.random.normal(loc=170,scale=10,size=1000)
plt.hist(data,bins=30,color='skyblue',edgecolor='black')
plt.xlabel('x axis')
plt.ylabel('y axis')
plt.title('Histogram Plot')
plt.show()
```
