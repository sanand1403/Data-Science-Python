```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def cost(w):
  return w ** 2
```

```python
def gradient(w):
  return 2 * w # Derivative of w^2
```

```python
# gradient descent function
def gradient_descent(start_w,learning_rate,epochs):
   w = start_w
   history = [w] # track updates
   for _ in range(epochs):
      grad = gradient(w)
      w = w - learning_rate * grad
      history.append(w)
   return history
```

```python
# start 4:3
start_w = 4.0
learning_rate = 0.1
epochs = 20

updates = gradient_descent(start_w,learning_rate,epochs)
print(updates[:5])
```

```python
# plot
plt.plot(range(epochs+1),[cost(w) for w in updates],marker='o')
plt.title("Gradient Descent")
plt.xlabel("Updates")
plt.ylabel("Cost (w^2)")
plt.grid(True)
plt.show()
```
