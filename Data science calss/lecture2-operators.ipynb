```python
# Arithmetic Operators (Math Tools)
x = 127
y = 45
print(x+y)
print(x-y)
print(x*y)
print(x/y)
print(x//y)
print(x%y)
print(x**3)
```

```python
# Comparison Operators (Checking Things)

print(23>45)
print(23<45)
print(23==45)
print(23!=45)
print(23<=45)
print(23>=45)
```

```python
# Logical Operators (Combining Checks)

print(23==45 and 56<67)
print(23==45 or 45>23)
print(not 25==25)
```

```python
# Assignment Operators (Saving Values)

x=10
print(x)
# x = x+10
x+=10
print(x)
x-=10
print(x)
x*=10
print(x)
x/=10
print(x)
```

```python
# Membership Operators (Checking Lists or Words)

data = "codewithanand"
print('w' in data)
data = "codewithanand"
print('w' not in data)
```

```python
# is 
# is not

data1 = 45
data2 = 67

print(data1 is data2)
print (data1 is not data2)
```
