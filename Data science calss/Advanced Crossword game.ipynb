```python
def crossword_game():
    print("=== Crossword Word Building Game ===")
    print("Rule: pick a letter from the current word.")
    print(" - The next word must START with that letter.")
    print(" - The length of the next word must equal the number of letters AFTER the chosen letter.")
    print(" - If you pick the last letter, the game ends.\n")

    word = input("Enter the first word: ").strip()
    if not word:
        print("No word entered. Exiting.")
        return

    while True:
        print(f"\nCurrent word: {word}")
        print("Positions:", " ".join(f"{i+1}:{ch}" for i, ch in enumerate(word)))

        choice = input("Choose a letter (or position number) from the word: ").strip()
        if not choice:
            print("Please enter a letter or a position number.")
            continue

        # allow choosing by position number
        if choice.isdigit():
            idx = int(choice) - 1
            if idx < 0 or idx >= len(word):
                print("Position out of range. Try again.")
                continue
            pos = idx
            letter = word[pos]
        else:
            # choose by letter (case-insensitive) - picks the first occurrence
            letter_lower = choice.lower()
            matches = [i for i, ch in enumerate(word) if ch.lower() == letter_lower]
            if not matches:
                print(f"Letter '{choice}' not found in the word. Try again.")
                continue
            pos = matches[0]
            letter = word[pos]
            if len(matches) > 1:
                print(f"Note: '{choice}' appears multiple times; using the occurrence at position {pos+1}.")

        # If chosen letter is the last letter, game ends
        if pos == len(word) - 1:
            print(f"You selected the last letter '{letter}'. Game Over!")
            break

        # Required length = letters after chosen one
        required_length = len(word) - pos - 1
        print(f"Next word must start with '{letter}' and have exactly {required_length} letters.")

        # prompt for next word until valid
        while True:
            new_word = input(f"Enter the next word: ").strip()
            if not new_word:
                print("Empty word. Try again.")
                continue
            if new_word[0].lower() != letter.lower():
                print(f"Invalid: the word must start with '{letter}'. Try again.")
                continue
            if len(new_word) != required_length:
                print(f"Invalid: the word must have exactly {required_length} letters. Try again.")
                continue
            # valid next word
            word = new_word
            break

if __name__ == "__main__":
    crossword_game()
```
